```python
train[['Prefix','Number']] = train['target'].str.extract(r'([A-Za-z]+)(\d+)', expand=True)
test[['Prefix','Number']] = test['target'].str.extract(r'([A-Za-z]+)(\d+)', expand=True)
from sklearn.preprocessing import LabelEncoder
ob=LabelEncoder()
train['Prefix']=ob.fit_transform(train['Prefix']).astype(int)
test['Prefix']=ob.fit_transform(test['Prefix']).astype(int)
```

```python
x_train = train.drop(['target','Prefix','Number'],axis=1)
y_train= train[['target','Prefix','Number']]
x_test = test.drop(['target','Prefix','Number'],axis=1)
y_test = test[['target']]
y_trainP=y_train['Prefix']
y_trainN=y_train['Number']
```

```python
"""from sklearn.linear_model import LinearRegression
l=LinearRegression()
l.fit(x_train,y_trainN)       ## as the numerical y isnt meaningful sequence so we apply classification model for this too
predN=l.predict(x_test)
predN=np.round(predN).astype(int)"""

from sklearn.neighbors import KNeighborsClassifier
obj=KNeighborsClassifier(n_neighbors=3)
obj.fit(x_train,y_trainN)
predN=obj.predict(x_test)

obj.fit(x_train,y_trainP)
predP=obj.predict(x_test)
predP=predP.astype(object)
predP[predP==0]="A"
predP[predP==1]="B"
```

```python
pred=predP+predN.astype(str)
pred
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(pred,y_test)
score
```

```python
from sklearn.cluster import KMeans
sse=[]
for k in range(1,20):
  km=KMeans(n_clusters=k)
  km.fit(x_train)
  sse.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.xticks(range(1,20))
plt.plot(range(1,20),sse)
```
